```python
np.random.seed(42)

data ={
    "product_id": range(1, 101),
    'product_name': [f'Product_{i}' for i in range(1, 101)],
    'category' :np.random.choice(['Electronics', 'Clothing', 'Home', 'Books'], 100),
    'units_sold' : np.random.poisson(lam=50, size=100),
    'price_per_unit': np.random.uniform(5.0, 500.0, size=100),
    'sale_date':pd.date_range(start='2023-01-01', periods=100, freq='D')    
        }
```

```python
data1 = pd.DataFrame(data)
data1.head(20)
```

```python
# Save the DataFrame as a CSV file
data1.to_csv('sales_data.csv', index=False)
```
